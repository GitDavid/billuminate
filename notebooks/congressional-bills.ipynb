### Aggregated bill data with: `data_agg.py`

```python
data_agg = pd.read_csv('agg_propublica_114hr.csv')
del data_agg['Unnamed: 0']
data_agg['bill_num'] = data_agg['bill_id'].apply(lambda x: int(x.split('hr')[1].split('-')[0]))
data_agg = data_agg.sort_values('bill_num')
data_agg = data_agg.reset_index(drop=True)
data_agg.head()
```

```python
data_agg.status.unique()
```

```python
data_agg.groupby('status').count()['bill_id']
```

```python
matplotlib.rcParams.update({'font.size': 14})
```

```python
fig, ax = plt.subplots(figsize=(8, 12))
color = matplotlib.cm.viridis_r(np.linspace(0, 1, len(data_agg.groupby('subjects_top_term'))))
data_agg.groupby('subjects_top_term').count()['bill_id'].sort_values().plot(kind='barh', ax=ax, color=color);
ax.grid(); ax.set_axisbelow(True); ax.set_ylabel('Bill Subject', size=26);
```
